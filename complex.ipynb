This is a more complete example. I often use close variant of the options used in this notebook for plots that I use in papers.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Note: If you use the following options, it will take slightly more time (~few seconds) to compile, but the math fonts looks nicer.

mpl.rc('figure')

# LaTex preamble
mpl.rcParams['text.usetex'] = True
mpl.rcParams[
    'text.latex.preamble'] = r'\usepackage[cm]{sfmath}\usepackage{amsmath}\usepackage{amssymb}'

# Use a sans-serif font with size 10
mpl.rcParams['font.family'] = 'sans-serif'
mpl.rcParams['font.sans-serif'] = 'CMU Sans Serif'
font_size = 10

fonts = {
    "axes.labelsize": font_size,
    "font.size": font_size,
    "legend.fontsize": font_size,
    "xtick.labelsize": font_size,
    "ytick.labelsize": font_size,
    "axes.titlesize": font_size,
}

mpl.rcParams.update(fonts)

# unobtrusive ticks and labels
axes_settings = {
    "axes.labelcolor": '#202020',
    "axes.spines.top": False,
    "axes.spines.right": False,
    "xtick.direction": "out",
    "ytick.direction": "out",
    "xtick.color": '#202020',
    "ytick.color": '#202020',
    "text.color": '#202020',
    "axes.edgecolor": '#202020',
}
mpl.rcParams.update(axes_settings)

mpl.rcParams['axes.grid'] = True
mpl.rcParams['grid.alpha'] = 0.5
mpl.rcParams['grid.color'] = 'grey'
mpl.rcParams['grid.linestyle'] = ':'
mpl.rcParams['grid.linewidth'] = 0.4
```

```python
x = np.arange(0,np.pi, 0.1)
sinx = np.sin(x)
cosx = np.cos(x)

# A single-column figure

fig, ax = plt.subplots(2, figsize=(3,4), constrained_layout=True)


# A plot

color1 = 'tab:red'
color2 = 'tab:gray'

ax[0].plot(x, sinx, label='$\sin(x)$', color=color1)
ax[0].plot(x, cosx, label='$\cos(x)$', color=color2)

ax[0].set_xlabel('$x$')
ax[0].set_xticks([0,np.pi/2, np.pi])
ax[0].set_xticklabels(['0', r'$\frac{\pi}{2}$', '$\pi$'])
ax[0].legend(handlelength=1,
           labelcolor='linecolor',
           frameon=False,
           loc='lower left')


# A histogram

# draw from normal distributions
mu = 1
sigma = 1

color3 = 'tab:green'
color4 = 'tab:blue'

ax[1].hist(np.random.normal(loc=mu, scale=sigma, size=10000), histtype='step', label='$\mathcal{N}(1,1)$', color=color3)
ax[1].hist(np.random.normal(loc=mu+2, scale=sigma, size=10000), histtype='step', label='$\mathcal{N}(3,0.5)$', color=color4)

ax[1].set_ylabel('Count')
# Create a separate legend to avoid ugly boxes for histogram legends
custom_lines = [
    mpl.lines.Line2D([0], [0], color=color3),
    mpl.lines.Line2D([0], [0], color=color4)
]
ax[1].legend(custom_lines, ['$\mathcal{N}(1,1)$', '$\mathcal{N}(3,1)$'],
           handlelength=1,
           labelcolor='linecolor',
           frameon=False,
           handletextpad=0.2, 
           loc='upper left',
           bbox_to_anchor=(0.7, 1))



# bbox_inches makes sure the texts at the sides of the figure are not cut while saving it as pdf
fig.savefig('./complex.pdf', bbox_inches='tight');
```
